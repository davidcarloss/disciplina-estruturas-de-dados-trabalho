```python
import matplotlib.pyplot as plt
import numpy as np
from main import MotorDeNegociacao
from gerador_testes import gerar_ordens_aleatorias, medir_tempo_processamento, medir_tempo_insercao_isolada
```

```python
volumes_teste = [1000, 2000, 4000, 8000, 16000, 32000, 64000] # O valor de 'n' dobra a cada passo
tempos_medios = []
rodadas_por_cenario = 3 # Número de vezes que cada teste é executado para obter um tempo médio mais preciso e evitar flutuações aleatórias de processamento
```

```python
print("Iniciando testes de performance...")

for n in volumes_teste:
    tempos_rodada = []
    
    for _ in range(rodadas_por_cenario):
        # Cria uma instância limpa do motor para cada rodada
        motor = MotorDeNegociacao()
        ordens = gerar_ordens_aleatorias(n)
        
        tempo_execucao = medir_tempo_processamento(motor, ordens)
        tempos_rodada.append(tempo_execucao)
        
    # Calcula o tempo médio para o 'n' atual
    tempo_medio = np.mean(tempos_rodada)
    tempos_medios.append(tempo_medio)
    print(f"Volume: {n} ordens -> Tempo Médio: {tempo_medio:.4f} segundos")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(volumes_teste, tempos_medios, marker='o', linestyle='-', color='b', linewidth=2)

plt.title('Performance do Motor de Negociação (Listas Lineares)', fontsize=14)
plt.xlabel('Volume de Ordens (n)', fontsize=12)
plt.ylabel('Tempo de Processamento (segundos)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Adicionando uma anotação estatística
plt.annotate('Comportamento Assintótico O(n²)', 
             xy=(volumes_teste[-1], tempos_medios[-1]), 
             xytext=(volumes_teste[-2], tempos_medios[-1]),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
# Célula: Análise Isolada da Inserção Ordenada O(n)
tempos_insercao = []

print("Iniciando teste isolado de Inserção na Lista Duplamente Encadeada...")

for n in volumes_teste:
    tempos_rodada_ins = []
    
    for _ in range(rodadas_por_cenario):
        ordens = gerar_ordens_aleatorias(n)
        tempo_exec = medir_tempo_insercao_isolada(ordens)
        tempos_rodada_ins.append(tempo_exec)
        
    tempo_medio_ins = np.mean(tempos_rodada_ins)
    tempos_insercao.append(tempo_medio_ins)
    print(f"Inserção de {n} ordens -> Tempo Médio: {tempo_medio_ins:.4f} s")

# Plotagem do Gráfico Isolado
plt.figure(figsize=(10, 6))
plt.plot(volumes_teste, tempos_insercao, marker='s', linestyle='-', color='r', linewidth=2)

plt.title('Custo da Inserção Ordenada (Lista Duplamente Encadeada)', fontsize=14)
plt.xlabel('Volume de Ordens Inseridas (n)', fontsize=12)
plt.ylabel('Tempo de Execução (segundos)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.annotate('Crescimento Quadrático Total\n(Devido a n inserções de custo médio O(n))', 
             xy=(volumes_teste[-1], tempos_insercao[-1]), 
             xytext=(volumes_teste[-2] * 0.8, tempos_insercao[-1] * 0.9),
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="r", lw=1))

plt.show()
```
